# pandas Operations and Plots

```python
import pandas as pd
import numpy as np

# Load the dataset
filename = '../data/Pokemon.csv'
PKMN = pd.read_csv(filename)
```

A series is a user facing class in `pandas`. Some methods include `sum`, `multiply`, and `add`.

```python
type(PKMN['Name'])
```

```python
table = pd.DataFrame({'A': [0.1, 0.2, 0.3], 'B': [10, 20, 30]})
table['B'].sum() # Summing values in a Series
```

```python
table['C'] = table['A'].add(table['B']) # Adding two Series
table
```

```python
charz_attack = PKMN.loc[(PKMN['Name']== 'Charizard'), ['Attack', 'Sp. Atk']]
charz_attack.values.sum() # Summing values in a DataFrame
```

```python
charz_attack['Total Attack'] = charz_attack['Attack'].add(charz_attack['Sp. Atk'])  # Adding two columns in a DataFrame
charz_attack
```

```python
charz_attack['Attack / Sp. Atk Ratio'] = charz_attack['Attack'].divide(charz_attack['Total Attack'])
charz_attack
```

```python
from sklearn import datasets
X = datasets.load_iris(as_frame=True)
df = X.frame
```

```python
df # Display the first few rows of the DataFrame
```

```python
df.plot(kind='hist') # Plotting a histogram of the DataFrame
df.plot(kind='hist', y=['sepal length (cm)', 'sepal width (cm)']) # Plotting specific columns
```

```python
df.plot.scatter(x='sepal length (cm)', y='sepal width (cm)') # Scatter plot of two columns
df.plot.scatter(x='sepal length (cm)', y='sepal width (cm)', grid=True) # Scatter plot with grid lines
```
